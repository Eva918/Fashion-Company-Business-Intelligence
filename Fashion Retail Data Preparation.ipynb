```python
import sys
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
```

```python
#7
# Load the dataset and select only the first 5000 rows
df = pd.read_csv("C:/Users/Evang/Downloads/mock_fashion_data_uk_us.csv").head(5000)
df
```

```python
scaler = MinMaxScaler()
df['Price'] = scaler.fit_transform(df[['Price']]) * 100  # Scale it to a numeric range, e.g., 0 to 100
df['Price'] = df['Price'].round(1)  # Round to one decimal place
# Round 'Rating' to one decimal place
# Round the 'Rating' column to the nearest whole number
df['Rating'] = df['Rating'].apply(lambda x: round(x))
# Define age groups and corresponding labels
bins = [0, 30, 50, 100]  # Define the bins for age ranges
labels = ['Young', 'Middle-aged', 'Senior']  # Define labels for each group

# Create a new column 'Age Group' based on the age bins
df['Age Group'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False)


# Display the result
print(df)
```

```python
data_info = df.info()

# Describe the dataset for numerical columns
description = df.describe()

data_info, description
```

```python
missing_values = df.isnull().sum()

# Display missing values
print(missing_values)
```

```python
# Convert the 'Rating' column to string type
df['Rating'] = df['Rating'].apply(str)

# Verify that the 'Rating' column is now of string type
print(df.dtypes['Rating'])
```
